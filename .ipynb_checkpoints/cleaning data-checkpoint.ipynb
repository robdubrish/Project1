```python
#import data found in https://www.kaggle.com/nickhould/craft-cans
beer_df = pd.read_csv('Resource/beers.csv')
breweries_df = pd.read_csv('Resource/breweries.csv')
beer_df.head()
```

```python
#there are 2410 beers in this list. We need to clean the data and remove the blank data.
beer_df = beer_df.dropna(axis=0, how='any')
beer_df.info()
```

```python
#clean data RE:breweries.csv
#rename unamed column to be able to merge data
breweries_df = breweries_df.rename(columns = {'Unnamed: 0':'brewery_id'})
breweries_df.head()
#there are 558 different breweries
```

```python
state = pd.DataFrame(breweries_df.groupby(['city'])['state'].count())
state_count = state.sort_values(by='state',ascending=False)
state_count.head()
```

```python
#find brewery count by city
city = pd.DataFrame(breweries_df.groupby(['state'])['city'].count())
city_count = city.sort_values(by='city',ascending=False)
city_count.head()
```

```python
#merge data to beers
merged = pd.merge(breweries_df,beer_df,on='brewery_id',suffixes=('_brewery','_beer'))
merged.head()
merged.info()
```

```python
brewerie_with_most = pd.DataFrame(merged.groupby(['name_brewery'])['name_beer'].count())
most_beers = brewerie_with_most.sort_values(by='name_beer',ascending=False)
most_beers.h
```
